# load data

```python
import pandas as pd
import numpy as np
```

```python
alltrain=pd.read_csv("train.csv",index_col=0,header=0)
alltrain
```

```python
alltest=pd.read_csv("test.csv",index_col=0,header=0)
alltest['SalePrice']=np.NaN
alltest
```

```python
alldata=alltrain.append(alltest)
alldata
```

```python
onehot2str={}
str_cols=[]
for c in alldata.columns:
    if str(alldata[c].dtype)=='object':
        str_cols.append(c)
        mapper={}
        idx=1
        for tag in alldata[c].unique():
            mapper[idx]=tag
            alldata.loc[alldata[c].values==tag,c]=idx
            idx+=1
            
        onehot2str[c]=mapper
print(str_cols)
```

```python
onehot2str
```

```python
alldata
```

# fill in missing values

## KNN


```python
from sklearn import neighbors
```

```python
tempdata=alldata.loc[:,~alldata.isnull().any().values]
n_neighbors=10
# 如果包含空值的列是object列，那么是离散的，应该用聚类，否则回归
for col in alldata.columns[alldata.isnull().any().values]:
    if col=="SalePrice":
        continue
    # 按照行切分，提取非nan的行用于训练
    notnan=alldata.loc[:,col].notna()
    isnan=alldata.loc[:,col].isna()
    if col in str_cols:
        # 聚类
        clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
        clf.fit(tempdata.loc[notnan], alldata.loc[notnan,col].astype(int))
        alldata.loc[isnan,col]=clf.predict(tempdata.loc[isnan])
    else:
        # 回归
        rgs = neighbors.KNeighborsRegressor(n_neighbors, weights='distance')
        rgs.fit(tempdata.loc[notnan], alldata.loc[notnan,col].astype(int))
        alldata.loc[isnan,col]=rgs.predict(tempdata.loc[isnan])
```

# split train and valid

```python
test=alldata.loc[alldata.loc[:,"SalePrice"].isna()]
train1=alldata.loc[~alldata.loc[:,"SalePrice"].isna()]
```

```python
sorted_train=train1.sort_values(["YrSold","MoSold"])
sorted_train
```

```python
valid=sorted_train.groupby(["YrSold","MoSold"]).sample(frac=0.1, random_state=2)
valid
```

```python
valid.groupby(["YrSold","MoSold"]).count()
```

```python
sorted_train.groupby(["YrSold","MoSold"]).count()
```

```python
train_idx=list(set(list(sorted_train.index)).difference(set(list(valid.index))))
len(train_idx)
```

```python
train=sorted_train.loc[train_idx].sort_values(by=["YrSold","MoSold"])
train
```

```python
set(list(train.index)).intersection(set(list(valid.index)))
```

```python
sorted_train.groupby(["YrSold","MoSold"]).mean().to_csv("mean_data.csv")
```
