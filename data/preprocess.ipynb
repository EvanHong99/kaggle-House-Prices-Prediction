# load data

```python
import pandas as pd
import numpy as np
from copy import deepcopy
from sklearn import neighbors
```

```python
alltrain=pd.read_csv("train.csv",index_col=0,header=0)
alltrain
```

```python
alltest=pd.read_csv("test.csv",index_col=0,header=0)
alltest['SalePrice']=np.NaN
alltest
```

```python
alldata=alltrain.append(alltest)
alldata
```

# 法1：在原数据上用onehot替代

```python
onehot2str={}
str_cols=["MSSubClass"]
for c in alldata.columns:
    # print(c,str(alldata[c].dtype))
    if str(alldata[c].dtype)=='object' or c in str_cols:
        str_cols.append(c)
        mapper={}
        mapper[0]=np.NaN
        idx=1
        for tag in alldata[c].unique():
            if pd.isna(tag):
                print("nan")
                continue
            mapper[idx]=tag
            alldata.loc[alldata[c].values==tag,c]=idx
            idx+=1

        onehot2str[c]=mapper
        alldata[c].fillna(0,inplace=True)
print(str_cols)
mapped_alldata=deepcopy(alldata)
alldata=alltrain.append(alltest)
```

```python
onehot2str
```

```python
alldata
```

## fill in missing values - KNN


```python
# 用于将上面map好的数据进行knn补全
tempdata=alldata.loc[:,~alldata.isna().any().values]
n_neighbors=10
# 如果包含空值的列是object列，那么是离散的，应该用聚类，否则回归
for col in alldata.columns[alldata.isnull().any().values]:
    if col=="SalePrice":
        continue
    # 按照行切分，提取非nan的行用于训练
    notnan=alldata.loc[:,col].notna()
    isnan=alldata.loc[:,col].isna()
    if col in str_cols:
        # 注意，这里选择用dummy来处理非数字的字段，而数字字段选择用knn来补全
        # 聚类
        clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
        clf.fit(tempdata.loc[notnan], alldata.loc[notnan,col].astype(int))
        alldata.loc[isnan,col]=clf.predict(tempdata.loc[isnan])
    else:
        # 回归
        print(col)
        rgs = neighbors.KNeighborsRegressor(n_neighbors, weights='distance')
        rgs.fit(tempdata.loc[notnan], alldata.loc[notnan,col])
        alldata.loc[isnan,col]=rgs.predict(tempdata.loc[isnan])
```

```python
# dummy法：对数字nan进行补全
tempdata=mapped_alldata.loc[:,~mapped_alldata.isna().any().values]
n_neighbors=10
# 如果包含空值的列是object列，那么是离散的，应该用聚类，否则回归
for col in alldata.columns[alldata.isnull().any().values]:
    if col=="SalePrice":
        continue
    # 按照行切分，提取非nan的行用于训练
    notnan=alldata.loc[:,col].notna()
    isnan=alldata.loc[:,col].isna()
    if col not in str_cols:
        # 回归
        print(col)
        rgs = neighbors.KNeighborsRegressor(n_neighbors, weights='distance')
        rgs.fit(tempdata.loc[notnan], alldata.loc[notnan,col])
        alldata.loc[isnan,col]=rgs.predict(tempdata.loc[isnan])
```

# 法2：新增dummy变量

将所有非数字的字段中的值视为虚拟变量（包括nan）

```python
# dummy_na=True将缺失值也当作合法的特征值并为其创建指示特征
alldata_dummy = pd.get_dummies(alldata.iloc[:,:-1], dummy_na=True,drop_first=True,columns=str_cols)
alldata_dummy['SalePrice']=alldata.iloc[:,-1]
print(alldata_dummy.shape) # (2919, 320)
alldata_dummy
```

```python
alldata_dummy.to_csv("./alldata_dummy.csv")
```

```python
alldata=deepcopy(alldata_dummy)
```
