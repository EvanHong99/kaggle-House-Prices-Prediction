```python
import pandas as pd
import numpy as np
from copy import deepcopy
from sklearn import neighbors
from sklearn import preprocessing
```

Import Data

```python
alltrain=pd.read_csv("train.csv",index_col=0,header=0)
alltest=pd.read_csv("test.csv",index_col=0,header=0)
```

```python
alltrain.head(5)
```

```python
alltest['SalePrice'] = np.NaN
alltest.head(5)
```

```python
alldata = alltrain.append(alltest)
alldata
```

Normalization Data

```python
alldata.dtypes
num_data = alldata.select_dtypes(exclude=object)
numerical = num_data.columns[0:36]
string_data = alldata.select_dtypes(include=object)
categorical = string_data.columns
for column in numerical:
    alldata[column]=((alldata[column]-alldata[column].min())/(alldata[column].max()-alldata[column].min()))*10
```

KNN fill NaN values

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
alldata[categorical] = alldata[categorical].apply(lambda series: pd.Series(
        LabelEncoder().fit_transform(series[series.notnull()]),
        index=series[series.notnull()].index
    ))
print(alldata)
imp_num = IterativeImputer(estimator=RandomForestRegressor(),
                               initial_strategy='mean',
                               max_iter=1, random_state=0)
imp_cat = IterativeImputer(estimator=RandomForestClassifier(), 
                               initial_strategy='most_frequent',
                               max_iter=1, random_state=0)
alldata[numerical] = imp_num.fit_transform(alldata[numerical])
alldata[categorical] = imp_cat.fit_transform(alldata[categorical])
```

```python
alldata
```

Mapping Category Columns

```python
alldata.to_csv("prepossess_data.csv")
```

```python
alltrain.index
prepossess_train_real = alldata.drop(alltest.index)

prepossess_train_real
prepossess_train=prepossess_train_real.sample(frac=0.8,random_state=200)
prepossess_valid=prepossess_train_real.drop(prepossess_train.index)
prepossess_train.to_csv("prepossess_train.csv")
prepossess_valid.to_csv("prepossess_valid.csv")
```
